```python
date =  sorted(list(set(df.starting_date)))[::-1]

yr = (np.arange(15)+1)*0.25
yr = np.append(yr, 3.91666667)

d = dict(zip(date, yr))
d
```

```python
def draw_without_stdev(measure):
    g = df.groupby(['starting_date'])[measure].apply(lambda x: (np.mean(x), np.std(x))).sort_index(ascending=False)
    for idx in g.index:
        print(f'({d[idx]}, {round(g.loc[idx][0]*100, 2)})')
              
def draw_with_stdev(measure):
    g = df.groupby(['starting_date'])[measure].apply(lambda x: (np.mean(x), np.std(x))).sort_index(ascending=False)
    for idx in g.index:
        print(f'({d[idx]}, {round(g.loc[idx][0]*100, 2)}) +- ({round(g.loc[idx][1]*100, 2)}, {round(g.loc[idx][1]*100, 2)})')

```

```python
draw_with_stdev('pr1')
```

```python
draw_with_stdev('pr2')
```

```python
draw_with_stdev('pr5')
```

```python
draw_with_stdev('pr10')
```

```python
draw_with_stdev('rev1')
```
