```python
df['ndim'] = df['dim']
```

```python
df.groupby(['ndim'])['f1','auc','pr1','pr2','pr5','pr10','re1','re2','re5','re10','rev1','rev2','rev5','rev10'].apply(lambda x: round(100 * np.mean(x), 2))
```

```python
df.groupby(['ndim'])['f1','auc','pr1','pr2','pr5','pr10','re1','re2','re5','re10','rev1','rev2','rev5','rev10'].apply(lambda x: round(100 * np.std(x), 2))
```

```python
def draw_without_stdev(measure):
    g = df.groupby(['ndim'])[measure].apply(lambda x: np.mean(x))
    for idx in g.index:
        print(f'({idx}, {round(g.loc[idx][0]*100, 2)})')
              
def draw_with_stdev(measure):
    g = df.groupby(['ndim'])[measure].apply(lambda x: (np.mean(x), np.std(x)))
    for idx in g.index:
        print(f'({idx}, {round(g.loc[idx][0]*100, 2)}) +- ({round(g.loc[idx][1]*100, 2)}, {round(g.loc[idx][1]*100, 2)})')
```

```python
# Precision
draw_with_stdev('pr1')
draw_with_stdev('pr2')
draw_with_stdev('pr5')
draw_with_stdev('pr10')
```

```python
# Revenue
draw_with_stdev('rev1')
draw_with_stdev('rev2')
draw_with_stdev('rev5')
draw_with_stdev('rev10')
```
