```python
def draw_without_stdev(measure, indices, pos):
    g = df.groupby(['ntree', 'ndepth'])[measure].apply(lambda x: np.mean(x))
    for idx in indices:
        print(f'({idx[pos]}, {round(g[idx]*100, 2)})')
              
def draw_with_stdev(measure, indices, pos):
    g = df.groupby(['ntree', 'ndepth'])[measure].apply(lambda x: (np.mean(x), np.std(x)))
    for idx in indices:
        print(f'({idx[pos]}, {round(g[idx][0]*100, 2)}) +- ({round(g.loc[idx][pos]*100, 2)}, {round(g.loc[idx][pos]*100, 2)})')
    
```

Changing number of trees

```python
# Precision
draw_with_stdev('pr1', [(25, 4), (50, 4), (100, 4), (200, 4), (400, 4)], 0) 
draw_with_stdev('pr2', [(25, 4), (50, 4), (100, 4), (200, 4), (400, 4)], 0) 
draw_with_stdev('pr5', [(25, 4), (50, 4), (100, 4), (200, 4), (400, 4)], 0) 
draw_with_stdev('pr10', [(25, 4), (50, 4), (100, 4), (200, 4), (400, 4)], 0) 
```

```python
# Revenue
draw_with_stdev('pr1', [(25, 4), (50, 4), (100, 4), (200, 4), (400, 4)], 0) 
draw_with_stdev('pr2', [(25, 4), (50, 4), (100, 4), (200, 4), (400, 4)], 0) 
draw_with_stdev('pr5', [(25, 4), (50, 4), (100, 4), (200, 4), (400, 4)], 0) 
draw_with_stdev('pr10', [(25, 4), (50, 4), (100, 4), (200, 4), (400, 4)], 0) 
```
