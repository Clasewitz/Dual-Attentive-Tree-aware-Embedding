```python
dfcls = df[df.model == 'DATE_CLS_13-01-01']
dfrev = df[df.model == 'DATE_REV_13-01-01']
```

```python
def draw_without_stdev(measure, df):
    g = df.groupby(['alpha'])[measure].apply(lambda x: (np.mean(x), np.std(x)))
    for idx in g.index:
        print(f'({idx}, {round(g.loc[idx][0]*100, 2)})')
              
def draw_with_stdev(measure,df):
    g = df.groupby(['alpha'])[measure].apply(lambda x: (np.mean(x), np.std(x)))
    for idx in g.index:
        print(f'({idx}, {round(g.loc[idx][0]*100, 2)}) +- ({round(g.loc[idx][1]*100, 2)}, {round(g.loc[idx][1]*100, 2)})')
```
